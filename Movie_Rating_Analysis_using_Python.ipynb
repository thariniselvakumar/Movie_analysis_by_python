```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("/content/movies.dat" , delimiter = "::" , engine= "python")
mov = movies.head()
mov
```

```python
movies.columns = ["ID","Title","Genre"]
print(movies.head())
```

```python
ratings = pd.read_csv("/content/ratings.dat", delimiter = "::" , engine = "python")
ratings.head()
```

```python
ratings.columns=["ID","Movies","Rating","Timestamp"]
ratings.head()
```

```python
data = pd.merge(movies , ratings , on=["ID","ID"])
data.head()
```

```python
import matplotlib.pyplot as plt

ratings = data["Rating"].value_counts()
numbers = ratings.index
quantity = ratings.values

plt.figure(figsize=(10,8))
plt.pie(quantity, labels=numbers, autopct='%1.1f%%')
plt.title('Ratings Distribution')
plt.show()
```
